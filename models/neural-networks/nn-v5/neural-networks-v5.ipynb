```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import os
from copy import deepcopy
from google.colab import drive
```

```python
DRIVE_PATH = os.path.join("./drive", "MyDrive", "cs3244-pg26")
PROCESSED_DATA_PATH = os.path.join("data_processed", "resale_flat_prices")
SEED = 42
```

```python
df = pd.read_csv(os.path.join(DRIVE_PATH, PROCESSED_DATA_PATH, "engineered_data.csv"))
df.head(10)
```

```python
from sklearn.impute import SimpleImputer

df["is_imputted_sora"] = np.where(df["avg_sora"].isnull(), 1, 0)

imp = SimpleImputer()
df["avg_sora"] = imp.fit_transform(df["avg_sora"].to_numpy().reshape(-1, 1))

df.drop(columns=["id", "sale_date", "address", "latitude_rad", "longitude_rad", "latitude", "longitude", "psm"], inplace=True)

df = pd.get_dummies(df, columns=["flat_type", "flat_model", "town"], dummy_na = True, dtype=np.uint8)
df.drop(columns=["town_nan", "flat_type_nan", "flat_model_nan"], inplace=True)

print(df.shape, '\n', df.columns)
display(df.head(3))
```

```python
lookback_years = 4
latest_month = df.sale_month.max()
test_months = list(range(latest_month - lookback_years * 12, latest_month + 1))
train_months = list(range(1, latest_month - lookback_years * 12))

df_train, df_test = deepcopy(df[df.sale_month.isin(train_months)]), deepcopy(df[df.sale_month.isin(test_months)])

X_train, X_test, y_train, y_test = df_train.drop('cpi_psm', axis=1), df_test.drop('cpi_psm', axis=1), df_train.cpi_psm, df_test.cpi_psm

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### Model design and training

```python
X_test.shape
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(57, activation="relu", input_shape=(57,)),
  tf.keras.layers.Dense(20, activation="relu"),
  tf.keras.layers.Dense(1, activation="linear")
])
```

```python
model.compile(loss="mse", optimizer="adam", metrics=["mse", "mae"])
model.summary()
```

```python
checkpoint_path = "model_checkpoints_final/cp-{epoch:04d}.ckpt"

batch_size = 1024

# Create a callback that saves the model's weights every 5 epochs
cp_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(DRIVE_PATH, checkpoint_path), 
    verbose=1, 
    save_weights_only=True,
    save_freq=5*batch_size)

model.save_weights(checkpoint_path.format(epoch=0))
```

```python
history = model.fit(X_train, y_train, batch_size = batch_size, epochs=400, validation_split=0.10, shuffle=True, callbacks=[cp_callback], verbose=1)
```

### Model analysis

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.subplot(111)
plt.plot(history.history["mae"][100:])
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.title("MAE vs Number of Epochs")
plt.show()
```

```python
plt.xlabel("Epoch")
plt.ylabel("RMSE")
plt.title("RMSE vs Number of Epochs")
plt.plot(np.sqrt(history.history["mse"][100:]))
plt.show()
```

```python
plt.subplot(111)
plt.xlabel("Epochs")
plt.ylabel("Validation MAE")
plt.title("Validation MAE vs Number of Epochs")
plt.plot(history.history["val_mae"])
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error, median_absolute_error
def produce_metrics(y_test, y_pred):
  mse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  rtwo = r2_score(y_test, y_pred)
  mape = mean_absolute_percentage_error(y_test, y_pred)
  mdape = median_absolute_error(y_test, y_pred)/np.median(y_test)
  return mse, mse**(0.5), mae, rtwo, mape, mdape
```

```python
mse, rmse, mae, rtwo, mape, mdape = produce_metrics(y_test, y_pred)
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R2: {rtwo}")
print(f"MAPE: {mape}")
print(f"MdAPE: {mdape}")
```
